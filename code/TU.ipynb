```python
import matplotlib.pyplot as plt
import pylab as pl

plt.figure(figsize=(7, 7)) 

x = ['Filter method', 'Wrapper method', 'Embedded method', 'Group Structure','Tree Structure']
y = [0.74,0.73,0.76,0.45,0.55]

pl.xticks(rotation=18)

plt.axis([-1, 5, 0.3, 0.8])

plt.grid(axis='y')
# 设置数字标签
for a, b in zip(x, y):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=13)
    
#plt.plot(x, y, color="b", linestyle="-", marker="*", linewidth=1.0)

plt.bar(x=x, height=y, color='indianred', alpha=0.8)

# 设置标题
plt.title("Forest-type-mapping Dataset Accuracy Rates")

plt.savefig('./1.jpg')
plt.show()
```

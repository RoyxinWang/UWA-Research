```python
import pandas as pd 
from sklearn.datasets import load_breast_cancer
def load_data(File):
    return pd.read_csv(File,header=None,sep = ',',index_col=False)

iris = load_breast_cancer()
```

```python
data, label = iris.data, iris.target
```

```python
from sklearn import feature_selection
import sys
import time
import numpy as np
from sklearn.datasets import load_iris
from scipy.sparse import rand
from skfeature.function.structure import tree_fs

start = time.clock()
#df = load_iris()
#y = df.target
#data = load_iris().data # shape=(150,4)

data, y = data,label
print(data.shape)

idx1 = np.array([[-1, -1, 1],[1, 1, np.sqrt(25)], [2, 2, np.sqrt(25)], [3, 3, np.sqrt(25)],[4, 4, np.sqrt(25)]]).T
#print(idx1)
idx2 = idx1.astype(int)
#print(idx2)

# simulate the dataset
X = data
# obtain the ground truth of the simulated dataset
#y = df.target
# specify the regularization parameter of regularization parameter of L2 norm for the non-overlapping group
z = 0.01

# perform feature selection and obtain the feature weight of all the features
w, obj, value_gamma = tree_fs.tree_fs(X, y, z, idx2, verbose=True)
print(w)

end = time.clock()
print(end-start)
```

```python
# remove the feature under 0.01
import pandas as pd
import numpy as np

#x=[0,2,3,5,6,7]
w[w<1.5e-05] = 0
w[w>=1.5e-05] = 1
#print(w)
x=[]
for i in range(0, len(w)):
    if(w[i]==0):
        x.append(i)


df = pd.DataFrame(data)
X_new = df.drop(df.columns[x],axis=1)
X_new.head(5)
```

```python
from sklearn.model_selection import train_test_split
# splitting the dataset - dataset still contains class label
X_train, X_test, Y_train, Y_test = train_test_split(X_new, label, test_size=0.3, random_state=0)

from sklearn.svm import SVC
svc = SVC().fit(X_train, Y_train)
svc_pre = svc.predict(X_test)
acc_svc = sum(svc_pre==Y_test)/len(Y_test)
print(acc_svc)
```
